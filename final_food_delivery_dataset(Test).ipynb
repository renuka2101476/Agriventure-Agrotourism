```python
user_totals = df.groupby('user_id')['total_amount'].sum()
count_users = (user_totals > 1000).sum()

if count_users < 500:
    q3 = "< 500"
elif count_users <= 1000:
    q3 = "500 – 1000"
elif count_users <= 2000:
    q3 = "1000 – 2000"
else:
    q3 = "> 2000"
print("Users spending > ₹1000:", q3)
```

```python
values = [0,3.5,4.0,4.5,5.0]
labels = ['3.0 – 3.5','3.6 – 4.0','4.1 – 4.5','4.6 – 5.0']
df['rating_range'] = pd.cut(df['rating'], bins=values, labels=labels, include_lowest=True)
q4 = df.groupby('rating_range', observed=False)['total_amount'].sum().idxmax()
print(" Highest revenue rating range:", q4)

q5 = df[df['membership']=='Gold'].groupby('city', observed=False)['total_amount'].mean().idxmax()
print("Gold highest avg order city:", q5)
```

```python
q6 = df.groupby('cuisine')['restaurant_id'].nunique().idxmin()
print(" Lowest restaurant count cuisine:", q6)

q7 = round((df[df['membership']=='Gold'].shape[0] / df.shape[0]) * 100)
print("% Orders by Gold members:", str(q7) + "%")
restaurant_stats = df.groupby('restaurant_name_y').agg(
    avg_order_value=('total_amount','mean'),
    order_count=('order_id','count')
)

q8 = restaurant_stats[restaurant_stats['order_count'] < 20] \
        .sort_values('avg_order_value', ascending=False) \
        .head(1).index[0]

print(" Highest AOV (<20 orders) restaurant:", q8)
```

```python
df['combo'] = df['membership'] + " + " + df['cuisine']
q9 = df.groupby('combo')['total_amount'].sum().idxmax()

print("Highest revenue combination:", q9)

df['quarter'] = df['order_date'].dt.to_period('Q')
q10 = df.groupby('quarter')['total_amount'].sum().idxmax()

print("Highest revenue quarter:", q10)
```
